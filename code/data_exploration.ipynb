```python
for i in range(min(1000, X_train.shape[1])):
    feat = X_train[:, i]
    print(f"Feature {i}: mean={np.mean(feat):.4f}, std={np.std(feat):.4f}, min={np.min(feat):.4f}, max={np.max(feat):.4f}")
```

```python
num_features = X_train.shape[1]
nan_count = 0
batch_size = 10000
for start in range(0, num_features, batch_size):
    end = min(start + batch_size, num_features)
    batch = X_train[:, start:end]
    nan_count += np.isnan(batch).sum()
    
print(f"NaNs in X_train: {nan_count}")

#no NaN is good
```
